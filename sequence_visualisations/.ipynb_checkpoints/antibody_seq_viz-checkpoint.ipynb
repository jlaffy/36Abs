<br><br>
Create function align(unaligned_df) to align data with the sequence alignments...
and gap_to_nan(df, gap_symbol) to replace a given gap symbol in a dataset with numpy NaN types: 

```python
def align(unaligned_df):
    aligned_df = unaligned_df.copy(deep=True)
    # for each sequence
    for seq in sequences.iterrows():
        seq_name = seq[0]
        #for each position in seqence
        for i, val in enumerate(seq[1]):
            #if there is a gap
            if val == '-':
                #get the unaligned values from here to the end, drop a nan
                slic = aligned_df.loc[seq_name, str(i+1):str(len(aligned_df.loc[seq_name])-1)]
                #make the value of the gap position a gap
                aligned_df.loc[seq_name, str(i+1)] = np.nan
                #replace the rest of the sequence with the values saved in slic
                for j, v in enumerate(slic):
                    aligned_df.loc[seq_name, str(i+j+2)] = v
    return aligned_df   
```

```python
def gap_to_nan(df, gap_symbol):
    for row in df.iterrows():
        for i, v in enumerate(row[1]):
            if v == gap_symbol:
                df.loc[row[0], str(i+1)] = np.nan
    return df
```
